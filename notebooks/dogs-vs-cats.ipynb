```python
"""
This notebook explores whether there are more dog lovers or cat lovers uploading photos of their pets onto Flickr
"""
```

```python
import flickrapi
from flickrapi import shorturl
import json

APIKEY = "800f28ae42a58a95f0005fb0fe022a78"

flickr = flickrapi.FlickrAPI(APIKEY)
```

```python
"""
Given a flickr api search query with format='json', returns the data in actual json format
flickr_search_response = flickr.photos_search(
        format='json', 
        ...
    )
"""
def get_data(flickr_search_response):
    json_string = flickr_search_response[14:-1]
    return json.loads(json_string)['photos']

"""
Returns the list of photos in results
"""
def get_photos(flickr_search_response):
    json_string = flickr_search_response[14:-1]
    return json.loads(json_string)['photos']['photo']

"""
Returns the count of results
"""
def get_count(flickr_search_response):
    json_string = flickr_search_response[14:-1]
    return json.loads(json_string)['photos']['total']
```

```python
tagged_dog = flickr.photos_search(
        tag_mode='any',
        tags='dog, puppy',
        license='7',
        format='json')

desc_dog = flickr.photos_search(
        text='dog',
        license='7',
        format='json')
```

```python
tagged_cat = flickr.photos_search(
        tag_mode='any',
        tags='cat, kitty, kitten',
        license='7',
        format='json')

desc_cat = flickr.photos_search(
        text='cat',
        license='7',
        format='json')
```

```python
"""
There are more users across the world uploading and tagging their dog photos on Flickr, compared to cat photos.
"""
```

```python
get_count(tagged_dog)
```

```python
get_count(tagged_cat)
```
